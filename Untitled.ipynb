```python
column_barchart = ['Time', 'Date', 'Merchant Group', 'Country of Transaction', 'Shipping Address', 'Country of Residence', 'Age', 'Bank']

for column in column_barchart:
    if data_fraud[column].dtype == 'object' or data_fraud[column].dtype == 'datetime64':
        total_count = data_fraud[column].dropna().count()
        top_values = data_fraud[column].value_counts().head(5)
        percentages = (top_values / total_count) * 100
        
        plt.figure(figsize=(11, 5))
        ax = sns.barplot(x=percentages, y=top_values.index, orient='h')
        plt.title(f'Percentage of {column}')
        plt.xlabel('Percentage')
        plt.ylabel(column)
        ax.xaxis.set_major_formatter(StrMethodFormatter('{x:,.2f}%'))
        
        for i, v in enumerate(percentages):
            ax.text(v, i, f'{v:.2f}%', color='black', ha='right')
        
        plt.show()
```

```python
column_piechart = ['Type of Card', 'Entry Mode', 'Type of Transaction', 'Gender']

for column in column_piechart:
    if data_fraud[column].dtype in ['object', 'int64']: 
        # Check if the column has a categorical or numerical data type
        
        # Count the occurrences of each category
        value_counts = data_fraud[column].value_counts()
        
        # Create a pie chart
        plt.figure(figsize=(8, 8))
        plt.pie(value_counts, labels=value_counts.index, autopct='%1.1f%%', startangle=140, pctdistance=0.85)
        plt.title(f'Pie chart for {column}')
        plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
        plt.show()
```

```python
# Create a copy of the original DataFrame
data_encoded_binary = data_3.copy()

# Iterate through each categorical column
for column in ['Country of Residence', 'Country of Transaction', 'Shipping Address','Type of Card','Entry Mode','Type of Transaction']:
    # Get unique categories in the column
    categories = data_encoded_binary[column].unique()
    
    # Create binary columns for each category
    for category in categories:
        # Create a new binary column with the category name
        data_encoded_binary[f'{column}_{category}'] = (data_encoded_binary[column] == category).astype(int)
    
    # Drop the original categorical column
    data_encoded_binary.drop(columns=[column], inplace=True)

# Display the first few rows of the modified DataFrame
print(data_encoded_binary.head())
```

```python
# Separate features and target variable, dropping specified columns
X = data_1.drop(columns=['Transaction ID', 'Fraud'])
y = data_1['Fraud']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

# RandomForestClassifier

```python
# Identify numeric and categorical columns
numeric_cols = X.select_dtypes(include=['int64', 'float64']).columns
object_cols = X.select_dtypes(include=['object']).columns

# Preprocessing steps
numeric_transformer = Pipeline(steps=[
    ('scaler', StandardScaler())
])

categorical_transformer = Pipeline(steps=[
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])

preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_cols),
        ('cat', categorical_transformer, object_cols)
    ])

# Define the Random Forest Classifier model
model = RandomForestClassifier(n_estimators=100, random_state=42)

# Create a pipeline that preprocesses data and fits the model
clf = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', model)
])

# Train the model
clf.fit(X_train, y_train)
```

```python
# Make predictions
predictions = clf.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, predictions)
print("Accuracy:", accuracy)
```

```python
# Confusion matrix
cm = confusion_matrix(y_test, predictions)

# Plot confusion matrix
plt.figure(figsize=(3, 3))
sns.heatmap(cm, annot=True, cmap='Greens', fmt='g')
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix')
plt.show()
```

```python
classification_rep = classification_report(y_test, predictions,digits=5)
print('Classification Report:')
print(classification_rep)
```

```python
from sklearn.metrics import cohen_kappa_score
```

```python
kappa_score = cohen_kappa_score(y_test, predictions)

print(f"Cohen's Kappa Score: {kappa_score:.4f}")
```

```python
from sklearn.metrics import roc_curve, roc_auc_score
y_probs = clf.predict_proba(X_test)[:, 1]

# Compute ROC curve
fpr, tpr, thresholds = roc_curve(y_test, y_probs)

# Compute AUC (Area Under the Curve)
roc_auc = roc_auc_score(y_test, y_probs)

# Plot ROC curve
plt.figure(figsize=(4,3))
plt.plot(fpr, tpr, color='blue', lw=2, label=f'ROC curve (area = {roc_auc:.4f})')
plt.plot([0, 1], [0, 1], color='gray', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc='lower right')
plt.show()
```
